```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# physical parameter
alpha = 1.0

# grid
Nx = 51
L = 1.0
dx = L / (Nx - 1)
x = np.linspace(0, L, Nx)

# time step (stable choice)

dt = 0.4 * dx**2 / alpha
r = alpha * dt / dx**2

print(f"Diffusion number r = {r:.3f}")
```

```python
# instable condition
dt=0.6 * dx**2 / alpha
r=alpha * dt / dx**2
print(f"Diffusion number r = {r:.3f}")
```

```python
T=np.zeros(Nx)
T_left=1
T_right=0
T[0], T[-1] = T_left, T_right
```

```python
#time integral
Nt=1000
snapshots=[]

for n in range(Nt):
  T_new=T.copy()
  for i in range(1,Nx-1):
    T_new[i]=T[i]+r*(T[i+1]-2*T[i]+T[i-1])
  T=T_new   #update

  if n in [0, 10, 50, 100, 300, 499]:
    snapshots.append(T.copy())
```

```python
#plot
for T_snap in snapshots:
  plt.plot(x,T_snap)
  plt.xlabel('x')
  plt.ylabel('T')
  plt.show()
```

```python
#checking time scale
time = Nt * dt
print(f"Total simulated time  {time:.2f}")
print(f"Diffusion time scale L^2/alpha = {L**2/alpha:.2f}")
```
